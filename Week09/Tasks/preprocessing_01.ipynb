```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
# Loading the dataset
data = pd.read_csv("data_01.csv")
data
```

```python
# Handling missing values
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(data.iloc[:, 1:3])
data.iloc[:, 1:3] = imputer.transform(data.iloc[:, 1:3])  

data
```

```python
# Apply OneHotEncoder on the first (Country) column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# [0] signifies the index of the column we are appliying the encoding on
column_transformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
data = pd.DataFrame(column_transformer.fit_transform(data))
data
```

```python
# Label encode the output

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data.iloc[:, -1] = label_encoder.fit_transform(data.iloc[:, -1])
data
```

```python
# Normalize the data so that no instance of data has dominance over another

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = pd.DataFrame(scaler.fit_transform(data))
data
```

```python
# Separate the data into inputs (x) and output (y)
inputs = data.iloc[:, :-1].values
output = data.iloc[:, -1].values
inputs, output
```

```python
from sklearn.model_selection import train_test_split
input_train, input_test, output_train, output_test = train_test_split(inputs, output, test_size = 0.2)
```
