```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('bank.csv', sep=';')
dataframe.head()
```

```python
def replace_marital_status(status):
  if status == "single":
    return 0
  
  return 1

dataframe["marital"] = dataframe["marital"].apply(replace_marital_status, 1)
dataframe.head()
```

```python
dataframe["housing"] = dataframe["housing"].map({ "yes": 1, "no": 0 })
dataframe.head()
```

```python
dataframe["loan"] = dataframe["loan"].replace({ "yes": 1, "no": 0 })
dataframe.head()
```

```python
# Get unique values of job
dataframe["job"].unique()
```

```python
dataframe["job"].replace({
  "unknown": np.nan,
  'unemployed': 0, 
  'services': 1, 
  'management': 2, 
  'blue-collar': 3,
  'self-employed': 4, 
  'technician': 5, 
  'entrepreneur': 6, 
  'admin.': 7, 
  'student': 8,
  'housemaid': 9, 
  'retired': 10
}, inplace=True)
dataframe.head()
```

```python
dataframe["education"].unique()
```

```python
dataframe["education"].replace({
  'primary': 2, 
  'secondary': 1, 
  'tertiary': 0, 
  'unknown': np.nan
}, inplace=True)
dataframe.head()
```

```python
dataframe["default"].replace({
  "yes": 1,
  "no": 0
}, inplace=True)
dataframe.head()
```

```python
dataframe["balance"].min()
```

```python
dataframe["balance"].max()
```

```python
def minimize(balance):
  minimum = dataframe["balance"].min()
  maximum = dataframe["balance"].max()
  return (balance - minimum) / (maximum - minimum)

dataframe["balance"] = dataframe["balance"].apply(minimize)
```

```python
dataframe.head()
```

```python
dataframe["contact"] = dataframe["contact"].replace({
  "unknown": np.nan,
  "telephone": 0,
  "cellular": 1
})
```

```python
dataframe["month"].unique()
```

```python
dataframe.month = dataframe.month.map({
  "jan": 1,
  "feb": 2,
  "mar": 3,
  "apr": 4,
  "may": 5,
  "jun": 6,
  "jul": 7,
  "aug": 8,
  "sep": 9,
  "oct": 10,
  "nov": 11,
  "dec": 12
})
```

```python
dataframe.head()
```

```python
dataframe["poutcome"].unique()
```

```python
dataframe["poutcome"].replace({
  "unknown": np.nan,
  "other": 0,
  "failure": 1,
  "success": 2
}, inplace=True)
```

```python
def minimize(pdays):
  minimum = dataframe["pdays"].min()
  maximum = dataframe["pdays"].max()
  return (pdays - minimum) / (maximum - minimum)

dataframe["pdays"] = dataframe["pdays"].apply(minimize)
```

```python
dataframe.head()
```

```python
dataframe["y"].replace({"no": 0, "yes": 1}, inplace=True)
```

```python
def minimize(duration):
  minimum = dataframe["duration"].min()
  maximum = dataframe["duration"].max()
  return (duration - minimum) / (maximum - minimum)

dataframe["duration"] = dataframe["duration"].apply(minimize)
```

```python
dataframe.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

column_transformer = ColumnTransformer([
  ("job", OneHotEncoder(), [1]),
  ("education", OneHotEncoder(), [3]),
  ("contact", OneHotEncoder(), [8]),
  ("month", OneHotEncoder(), [10]),
  ("poutcome", OneHotEncoder(), [15]),
], remainder="passthrough")
dataframe = column_transformer.fit_transform(dataframe)
```

```python
dataframe = pd.DataFrame(dataframe)
dataframe.head()
```

```python
from sklearn.model_selection import train_test_split

X = dataframe.iloc[:, :-1].values
y = dataframe.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```
